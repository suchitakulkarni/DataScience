```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from math import radians, sin, cos, sqrt, atan2
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.compose import TransformedTargetRegressor
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
# Constants
C = 3e8  # speed of light in vacuum (m/s)
FIBER_SPEED = 2e8  # ~2/3 speed of light in optical fiber
```

```python
# Function to compute great-circle distance (haversine formula)
def haversine(lat1, lon1, lat2, lon2):
    R = 6371e3  # Earth radius in meters
    phi1, lambda1 = radians(lat1), radians(lon1)
    phi2, lambda2 = radians(lat2), radians(lon2)
    dphi = phi2 - phi1
    dlambda = lambda2 - lambda1

    a = sin(dphi/2)**2 + cos(phi1) * cos(phi2) * sin(dlambda/2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    return R * c  # in meters

# Sample city coordinates (lat, lon)
cities = {
    'New York': (40.7128, -74.0060),
    'London': (51.5074, -0.1278),
    'Tokyo': (35.6895, 139.6917),
    'Sydney': (-33.8688, 151.2093),
    'San Francisco': (37.7749, -122.4194),
    'Paris': (48.8566, 2.3522),
    'Singapore': (1.3521, 103.8198),
    'Dubai': (25.2048, 55.2708),
    'Cape Town': (-33.9249, 18.4241),
    'Toronto': (43.6532, -79.3832),
    'Berlin': (52.5200, 13.4050),
    'Moscow': (55.7558, 37.6173),
    'Beijing': (39.9042, 116.4074),
    'Delhi': (28.6139, 77.2090),
    'Los Angeles': (34.0522, -118.2437),
    'Mexico City': (19.4326, -99.1332),
    'Buenos Aires': (-34.6037, -58.3816),
    'Johannesburg': (-26.2041, 28.0473),
    'Seoul': (37.5665, 126.9780),
    'Madrid': (40.4168, -3.7038),
    'Rome': (41.9028, 12.4964),
    'Istanbul': (41.0082, 28.9784),
    'Bangkok': (13.7563, 100.5018),
    'Jakarta': (-6.2088, 106.8456),
    'Cairo': (30.0444, 31.2357),
    'Athens': (37.9838, 23.7275),
    'Lima': (-12.0464, -77.0428),
    'Kuala Lumpur': (3.1390, 101.6869),
    'Hanoi': (21.0278, 105.8342),
    'Manila': (14.5995, 120.9842),
    'Santiago': (-33.4489, -70.6693),
    'Tehran': (35.6892, 51.3890),
    'Riyadh': (24.7136, 46.6753),
    'Doha': (25.276987, 51.520008),
    'Zurich': (47.3769, 8.5417),
    'Vienna': (48.2082, 16.3738),
    'Oslo': (59.9139, 10.7522),
    'Stockholm': (59.3293, 18.0686),
    'Helsinki': (60.1695, 24.9354),
    'Warsaw': (52.2297, 21.0122),
    'Prague': (50.0755, 14.4378),
    'Budapest': (47.4979, 19.0402),
    'Lisbon': (38.7169, -9.1399),
    'Brussels': (50.8503, 4.3517),
    'Amsterdam': (52.3676, 4.9041),
    'Montreal': (45.5017, -73.5673),
    'Vancouver': (49.2827, -123.1207),
    'Chicago': (41.8781, -87.6298),
    'Houston': (29.7604, -95.3698),
    'Atlanta': (33.7490, -84.3880),
    'Boston': (42.3601, -71.0589),
    'Miami': (25.7617, -80.1918),
    'Seattle': (47.6062, -122.3321),
    'Dallas': (32.7767, -96.7970),
    'Phoenix': (33.4484, -112.0740),
    'Philadelphia': (39.9526, -75.1652),
    'Detroit': (42.3314, -83.0458),
    'Denver': (39.7392, -104.9903),
    'Bangalore': (12.9716, 77.5946),
    'Chennai': (13.0827, 80.2707),
    'Hyderabad': (17.3850, 78.4867),
    'Ahmedabad': (23.0225, 72.5714),
    'Pune': (18.5204, 73.8567),
    'Karachi': (24.8607, 67.0011),
    'Lagos': (6.5244, 3.3792),
    'Nairobi': (-1.2921, 36.8219),
    'Casablanca': (33.5731, -7.5898),
    'Algiers': (36.7538, 3.0588),
    'Tunis': (36.8065, 10.1815),
    'Havana': (23.1136, -82.3666),
    'Caracas': (10.4806, -66.9036),
    'Bogotá': (4.7110, -74.0721),
    'Medellín': (6.2442, -75.5812),
    'Brasília': (-15.8267, -47.9218),
    'São Paulo': (-23.5505, -46.6333),
    'Rio de Janeiro': (-22.9068, -43.1729),
    'Recife': (-8.0476, -34.8770),
    'Porto Alegre': (-30.0346, -51.2177),
    'Quito': (-0.1807, -78.4678),
    'La Paz': (-16.5000, -68.1500),
    'Montevideo': (-34.9011, -56.1645),
    'Hobart': (-42.8821, 147.3272),
    'Auckland': (-36.8485, 174.7633),
    'Wellington': (-41.2865, 174.7762),
    'Suva': (-18.1248, 178.4501),
    'Reykjavik': (64.1265, -21.8174),
    'Tallinn': (59.4370, 24.7536),
    'Riga': (56.9496, 24.1052),
    'Vilnius': (54.6872, 25.2797),
    'Belgrade': (44.7866, 20.4489),
    'Zagreb': (45.8150, 15.9819),
    'Sarajevo': (43.8563, 18.4131),
    'Skopje': (41.9981, 21.4254),
    'Podgorica': (42.4304, 19.2594),
    'Tbilisi': (41.7151, 44.8271),
    'Yerevan': (40.1792, 44.4991),
    'Baku': (40.4093, 49.8671),
    'Islamabad': (33.6844, 73.0479),
    'Kathmandu': (27.7172, 85.3240),
    'Ulaanbaatar': (47.8864, 106.9057),
    'Astana': (51.1605, 71.4704)
}
print(len(cities))
# Generate city pairs (bi-directional)
pairs = []
for city_a in cities:
    for city_b in cities:
        if city_a != city_b:
            pairs.append((city_a, city_b))
```

```python
np.random.seed(42)
data = []
samples_per_pair = 2  # adjust as needed
for city_a, city_b in pairs:
    lat1, lon1 = cities[city_a]
    lat2, lon2 = cities[city_b]

    # Step 1: Geodesic distance
    dist_m = haversine(lat1, lon1, lat2, lon2)

    # Step 2: Stretch fiber distance to simulate real-world detours
    stretch_factor = np.random.normal(loc=1.3, scale=0.1)  # ~30% longer on average
    fiber_dist_m = dist_m * stretch_factor

    # Step 3: Physics-informed min latency using the *stretched* path
    min_latency = fiber_dist_m / FIBER_SPEED  # seconds
    min_latency_ms = min_latency * 1000

    for _ in range(samples_per_pair):
        # Instrumental noise
        noise_ms = np.random.normal(loc=0, scale=2)

        # Optional anomaly injection
        is_anomaly = np.random.rand() < 0.01
        if is_anomaly:
            anomaly_type = np.random.choice(["positive", "negative"], p=[0.8, 0.2])
            if anomaly_type == "positive":
                noise_ms += np.random.choice([10, 20, 50])
            else:
                noise_ms -= np.random.choice([2, 5, 8])

        measured_latency_ms = min_latency_ms + noise_ms

        data.append({
            'source': city_a,
            'destination': city_b,
            'geo_distance_km': dist_m / 1000,          # geodesic
            'fiber_distance_km': fiber_dist_m / 1000,  # actual fiber path
            'min_latency_ms': min_latency_ms,
            'measured_latency_ms': measured_latency_ms,
            'noise_ms': noise_ms,
            'is_anomaly': int(is_anomaly)
        })
df = pd.DataFrame(data)
df.to_csv('latency_simulated_data.dat')
```

```python
# Preview
print(df.head())

# Plot
plt.figure(figsize=(6, 6))
plt.scatter(df['fiber_distance_km'], df['measured_latency_ms'], label='Measured latency', alpha=0.7)
plt.plot(df['fiber_distance_km'], df['min_latency_ms'], 'r--', label='Light-speed latency (fiber)')

plt.xlabel('Distance (km)')
plt.ylabel('Latency (ms)')
plt.title('Synthetic Intercontinental Latency (Physics-Informed)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
feature_columns = ['fiber_distance_km'] #, 'hop_count', 'bandwidth_mbps']  # example
target_column1 = 'measured_latency_ms'
target_column2 = 'min_latency_ms'

# Data extraction
X = df[feature_columns].values
y = df[[target_column]].values # provides a 2D array which works well with Linear Regression
#y = df[[target_column1]].values - df[[target_column2]].values # provides a 2D array which works well with Linear Regression
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
pipeline = Pipeline([
                    ("scaler", StandardScaler()),
                    ("regressor", LinearRegression())
                    ])

model = TransformedTargetRegressor(
                                regressor=pipeline,
                                transformer=StandardScaler()
                                )


# Perform 5-fold cross-validation
scores = cross_val_score(model, X, y, cv=5, scoring="r2")

print("Cross-validated R² scores:", scores)
print("Mean R² score:", scores.mean())

model.fit(X, y)
y_pred = model.predict(X)
```

```python
# Plot
fig, ax = plt.subplots(figsize=(5, 5))
ax.scatter(y, y_pred, label='Linear regression prediction', alpha=0.7)
ax.plot([0, 1], [0, 1], transform=ax.transAxes)
ax.set_xlabel('Actual')
ax.set_ylabel('Predicted')
#plt.title('Synthetic Intercontinental Latency (Physics-Informed)')
ax.legend()
ax.grid(True)
plt.tight_layout()
plt.show()
```

# Residuals

```python
residuals = y.flatten() - y_pred.flatten()

# Assume residuals is a 1D NumPy array
centered_residuals = residuals - np.mean(residuals)
std = np.std(centered_residuals)
z_score = centered_residuals/std
anomaly_threshold = 2
anomalies_detected = np.abs(z_score) > anomaly_threshold
print(f"Detected {np.sum(anomalies_detected)} out of {len(residuals)} samples")
print(f"True anomalies {np.sum(df['is_anomaly'].tolist())}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Plot histogram
plt.figure(figsize=(8, 5))
#n, bins, patches = plt.hist(centered_residuals,density=True, bins=50, color='lightgray', edgecolor='black', label='Residuals')
n, bins, patches = plt.hist(centered_residuals, bins=50, color='lightgray', edgecolor='black', label='Residuals')

# Shade regions beyond ±2σ
for i in range(len(bins)-1):
    if bins[i] < -2*std or bins[i+1] > 2*std:
        patches[i].set_facecolor('red')

# Mark ±2σ lines
plt.axvline(-2*std, color='red', linestyle='--', linewidth=1)
plt.axvline(2*std, color='red', linestyle='--', linewidth=1)

# Labels and legend
plt.xlabel('Centered Residual (ms)')
plt.ylabel('Frequency')
plt.title('Residual Distribution with ±2σ Anomaly Regions')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
anomalies_label = df['is_anomaly'].values.astype(bool) 
# Categorize outcomes
true_positives = anomalies_detected & anomalies_label         # correctly detected
false_positives = anomalies_detected & ~anomalies_label       # falsely flagged
false_negatives = ~anomalies_detected & anomalies_label       # missed anomalies

plt.figure(figsize=(10, 6))

# All residuals (light gray)
plt.scatter(X, residuals, color='lightgray', label="All data", s=20)

# True Positives: correctly detected anomalies
plt.scatter(X[true_positives], residuals[true_positives],
            color='green', label="True Positives", s=50, marker='o')

# False Positives: flagged but not real anomalies
plt.scatter(X[false_positives], residuals[false_positives],
            color='orange', label="False Positives", s=50, marker='x')

# False Negatives: real anomalies not detected
plt.scatter(X[false_negatives], residuals[false_negatives],
            color='red', label="False Negatives", s=50, marker='^')

plt.xlabel("Distance (km)")
plt.ylabel("Residual (ms)")
plt.title("Detected vs True Anomalies")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df['instrumental_delay_ms'] = df['measured_latency_ms'] - df['min_latency_ms']
X = df[['fiber_distance_km']]
y = df[['instrumental_delay_ms']]
model_residual = LinearRegression()
```

```python
model_residual.fit(X, y)
df['predicted_latency_ms'] = df['min_latency_ms'] + model_residual.predict(X).flatten()
```

```python
residuals = y.values.flatten() - model_residual.predict(X).flatten()
mean_residuals = np.mean(residuals)
std_residuals = np.std(residuals)
z_score = (residuals - mean_residuals)/std_residuals
```

```python
plt.scatter(X, residuals)
plt.axhline(np.mean(residuals), color='red', linestyle='--')
```

```python
anomaly_threshold = 2
anomalies = np.abs(z_score) > anomaly_threshold
print(f"Detected {np.sum(anomalies)} out of {len(residuals)} samples")
```
