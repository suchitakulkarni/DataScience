```python
import numpy as np
import pandas as pd

np.random.seed(42)

#!/usr/bin/env python
# coding: utf-8

# # Latency Prediction: Physics vs Data-Driven Models
# 
# This notebook explores how purely data-driven models for latency prediction can learn biased baselines, how physics-informed models provide better anomaly detection, and how hybrid approaches combine the best of both worlds.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import f1_score
from sklearn.model_selection import train_test_split
from sklearn.utils import shuffle
```

```python

# Constants
C = 3e8  # speed of light in vacuum (m/s)
FIBER_SPEED = 2e8  # speed of light in fiber (m/s)
N = 3000

# Helper: Haversine distance in km
def haversine(lat1, lon1, lat2, lon2):
    R = 6371  # Earth radius km
    dlat = np.radians(lat2 - lat1)
    dlon = np.radians(lon2 - lon1)
    a = np.sin(dlat/2)**2 + np.cos(np.radians(lat1)) * np.cos(np.radians(lat2)) * np.sin(dlon/2)**2
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1-a))
    return R * c

# Generate random city pairs on Earth
lats1 = np.random.uniform(-60, 60, N)
lons1 = np.random.uniform(-180, 180, N)
lats2 = np.random.uniform(-60, 60, N)
lons2 = np.random.uniform(-180, 180, N)

geo_distance_km = haversine(lats1, lons1, lats2, lons2)

# Fiber distance slightly longer than geo (5%-30% longer)
fiber_distance_km = geo_distance_km * (1 + np.random.uniform(0.05, 0.3, N))

# Ideal latency (ms)
ideal_latency_ms = (fiber_distance_km * 1000) / FIBER_SPEED * 1000

# Systematic biases:
# Region-specific equipment delays (in ms)
def region_delay(lat, lon):
    # Example: +5ms delay if near some region (e.g., lat > 30 and lon < -50)
    return np.where((lat > 30) & (lon < -50), 5, 0)

region_delay_1 = region_delay(lats1, lons1)
region_delay_2 = region_delay(lats2, lons2)
region_bias = (region_delay_1 + region_delay_2) / 2

# Time-based congestion bias: sinusoidal over day fraction
time_of_day = np.random.uniform(0, 24, N)
congestion_bias = 3 * np.sin((time_of_day / 24) * 2 * np.pi)

# Measurement noise ~ N(0, 2ms)
noise = np.random.normal(0, 2, N)

# Measured latency (ms)
measured_latency_ms = ideal_latency_ms + region_bias + congestion_bias + noise

# Introduce anomalies: 5% of data
anomaly_indices = np.random.choice(N, size=int(0.05 * N), replace=False)
# Anomalies: latency > 1.5x physics + extra noise
measured_latency_ms[anomaly_indices] = measured_latency_ms[anomaly_indices] * np.random.uniform(1.6, 3, len(anomaly_indices)) + np.random.normal(10, 5, len(anomaly_indices))

# Label anomalies
is_anomaly = np.zeros(N, dtype=bool)
is_anomaly[anomaly_indices] = True

# Create DataFrame
df = pd.DataFrame({
    'lat1': lats1,
    'lon1': lons1,
    'lat2': lats2,
    'lon2': lons2,
    'geo_distance_km': geo_distance_km,
    'fiber_distance_km': fiber_distance_km,
    'measured_latency_ms': measured_latency_ms,
    'is_anomaly': is_anomaly
})

# Save to CSV
df.to_csv('latency_simulated_data.dat', index=False)
print("Simulation data saved to 'latency_simulated_data.dat'")
```

```python
import numpy as np
import pandas as pd

np.random.seed(42)

# Constants
FIBER_SPEED = 2e8  # speed of light in fiber (m/s)
N = 3000

# Helper: Haversine formula to calculate geo distance (in km)
def haversine(lat1, lon1, lat2, lon2):
    R = 6371  # Earth radius in km
    dlat = np.radians(lat2 - lat1)
    dlon = np.radians(lon2 - lon1)
    a = np.sin(dlat/2)**2 + np.cos(np.radians(lat1)) * np.cos(np.radians(lat2)) * np.sin(dlon/2)**2
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1-a))
    return R * c

# Randomly generate coordinates for city pairs
lat1 = np.random.uniform(-60, 60, N)
lon1 = np.random.uniform(-180, 180, N)
lat2 = np.random.uniform(-60, 60, N)
lon2 = np.random.uniform(-180, 180, N)

# Calculate geodesic (great-circle) distance
geo_distance_km = haversine(lat1, lon1, lat2, lon2)
stretch_factor = 1
# Physics-based ideal latency (lower bound)
ideal_latency_ms = stretch_factor*geo_distance_km * 1000 / FIBER_SPEED * 1000  # (km → m → s → ms)

# Regional delays (based on source coordinates)
def region_delay(lat, lon):
    return np.where((lat > 30) & (lon < -50), 5, 0)

region_bias_src = region_delay(lat1, lon1)
region_bias_dst = region_delay(lat2, lon2)
region_bias = (region_bias_src + region_bias_dst) / 2

# Simulate congestion effects (time of day bias)
time_of_day = np.random.uniform(0, 24, N)
congestion_bias = 3 * np.sin((time_of_day / 24) * 2 * np.pi)

# Add measurement noise
measurement_noise = np.random.normal(0, 1.5, N)

# Construct measured latency
measured_latency_ms = ideal_latency_ms + region_bias + congestion_bias + measurement_noise

# Add anomalies: latency > 1.5× physics + extra noise
is_anomaly = np.zeros(N, dtype=bool)
anomaly_indices = np.random.choice(N, size=int(0.05 * N), replace=False)
measured_latency_ms[anomaly_indices] = (
    ideal_latency_ms[anomaly_indices] * np.random.uniform(1.6, 2.5, len(anomaly_indices)) +
    np.random.normal(10, 5, len(anomaly_indices))
)
is_anomaly[anomaly_indices] = True

# Create dataframe
df = pd.DataFrame({
    'lat1': lat1,
    'lon1': lon1,
    'lat2': lat2,
    'lon2': lon2,
    'geo_distance_km': geo_distance_km,
    'measured_latency_ms': measured_latency_ms,
    'ideal_latency_ms': ideal_latency_ms,
    'is_anomaly': is_anomaly
})

# Save to file
df.to_csv('latency_simulated_data.dat', index=False)
print("Saved 'latency_simulated_data.dat' — with no fiber distances.")
```

```python
# Constants
C = 3e8  # Speed of light in vacuum (m/s)
FIBER_SPEED = 2e8  # ~2/3 speed of light in fiber (m/s)

# Load data
df = pd.read_csv('latency_simulated_data.dat')
df['ideal_latency_ms'] = (df['geo_distance_km'] * 1000) / FIBER_SPEED * 1000
```

```python
# -- 1. Show that Data-Driven Alone Learns the Wrong Baseline --

print("1. Data-Driven Baseline")

# Fit LinearRegression on observed fiber distance only
#X = df[['fiber_distance_km']].values
X = df[['geo_distance_km']].values
y = df['measured_latency_ms'].values

model_dd = LinearRegression()
model_dd.fit(X, y)
slope_dd = model_dd.coef_[0]
intercept_dd = model_dd.intercept_

print(f"Data-driven model slope: {slope_dd:.4f} ms/km")
print(f"Data-driven model intercept: {intercept_dd:.4f} ms")

# Compare to physics slope (ms/km)
physics_slope = 1000 / FIBER_SPEED * 1000  # ms/km (distance in km to ms)
print(f"Physics slope (ideal): {physics_slope:.4f} ms/km")

# Plot predicted latency by data-driven model vs true physics latency
plt.figure(figsize=(8,5))
#plt.scatter(df['fiber_distance_km'], df['measured_latency_ms'], alpha=0.4, label='Measured Latency')
#plt.plot(df['fiber_distance_km'], model_dd.predict(X), color='orange', label='Data-Driven LR fit')
#plt.scatter(df['fiber_distance_km'], df['ideal_latency_ms'], color='gray', label='Physics Ideal Latency')

plt.scatter(df['geo_distance_km'], df['measured_latency_ms'], alpha=0.4, label='Measured Latency')
plt.plot(df['geo_distance_km'], model_dd.predict(X), color='orange', label='Data-Driven LR fit')
plt.scatter(df['geo_distance_km'], df['ideal_latency_ms'], color='gray', label='Physics Ideal Latency')
plt.xlabel("Fiber Distance (km)")
plt.ylabel("Latency (ms)")
plt.title("Data-Driven vs Physics Ideal Latency")
plt.legend()
plt.grid()
plt.show()

# Plot residual error vs distance for data-driven model
residuals_dd = y - model_dd.predict(X)
plt.figure(figsize=(8,5))
#plt.scatter(df['fiber_distance_km'], residuals_dd, alpha=0.5)
plt.scatter(df['geo_distance_km'], residuals_dd, alpha=0.5)
plt.axhline(0, color='black', linestyle='--')
plt.xlabel("Fiber Distance (km)")
plt.ylabel("Residual Error (ms)")
plt.title("Data-Driven Model Residuals vs Distance")
plt.grid()
plt.show()

print("""
Observations:
- Data-driven model underestimates latency at long distances (residuals negative).
- Overestimates latency at short distances (residuals positive).
- Systematic bias due to routing shortcuts and regional complexity in training data.
- Data-driven learns biases, not pure physics.
""")
```

```python
# -- 2. Physics-Informed Residuals for Anomaly Detection --

print("\n2. Physics-Informed Residual Anomaly Detection")

# Physics prediction baseline
physics_pred = df['ideal_latency_ms'].values
residuals_phys = y - physics_pred

# Data-driven residuals already computed (residuals_dd)

# Anomaly labels
labels = df['is_anomaly'].values.astype(bool)

# Function to detect anomalies using z-score on residuals
def detect_anomalies(residuals, threshold=2):
    mean = np.mean(residuals)
    std = np.std(residuals)
    z = (residuals - mean) / std
    return np.abs(z) > threshold

# Detect anomalies
anomalies_phys = detect_anomalies(residuals_phys)
anomalies_dd = detect_anomalies(residuals_dd)

# Compute F1 scores
f1_phys = f1_score(labels, anomalies_phys)
f1_dd = f1_score(labels, anomalies_dd)

print(f"F1 Score Physics-informed residuals: {f1_phys:.3f}")
print(f"F1 Score Data-driven residuals: {f1_dd:.3f}")

# Plot residuals for physics model
plt.figure(figsize=(8,4))
plt.hist(residuals_phys, bins=50, alpha=0.7, label='Physics Residuals')
plt.axvline(np.mean(residuals_phys) + 2*np.std(residuals_phys), color='r', linestyle='--', label='±2 std dev')
plt.axvline(np.mean(residuals_phys) - 2*np.std(residuals_phys), color='r', linestyle='--')
plt.title("Physics-Informed Residuals Distribution")
plt.legend()
plt.grid()
plt.show()

# Plot residuals for data-driven model
plt.figure(figsize=(8,4))
plt.hist(residuals_dd, bins=50, alpha=0.7, label='Data-Driven Residuals', color='orange')
plt.axvline(np.mean(residuals_dd) + 2*np.std(residuals_dd), color='r', linestyle='--', label='±2 std dev')
plt.axvline(np.mean(residuals_dd) - 2*np.std(residuals_dd), color='r', linestyle='--')
plt.title("Data-Driven Residuals Distribution")
plt.legend()
plt.grid()
plt.show()

print("""
Physics-informed residuals have lower variance and yield better anomaly detection performance (higher F1).
Physics residuals anchor on a true baseline, unlike data-driven residuals which absorb biases.
""")
```

```python
# -- 3. Robustness to Biased Training Data --

print("\n3. Robustness to Biased Training Data")

# Simulate training sets biased toward short distances only (e.g. < 2000 km)
#biased_train_mask = df['fiber_distance_km'] < 2000
biased_train_mask = df['geo_distance_km'] < 2000
biased_train = df[biased_train_mask]
balanced_test = df[~biased_train_mask]

# Train data-driven model on biased short-distance training set
#X_biased = biased_train[['fiber_distance_km']].values
X_biased = biased_train[['geo_distance_km']].values
y_biased = biased_train['measured_latency_ms'].values

model_biased = LinearRegression()
model_biased.fit(X_biased, y_biased)

# Evaluate on balanced test set
#X_test = balanced_test[['fiber_distance_km']].values
X_test = balanced_test[['geo_distance_km']].values
y_test = balanced_test['measured_latency_ms'].values

y_pred_biased = model_biased.predict(X_test)

# Baseline physics predictions on test set
y_phys_test = balanced_test['ideal_latency_ms'].values

# Compute mean absolute errors on test set
mae_biased = np.mean(np.abs(y_pred_biased - y_test))
mae_phys = np.mean(np.abs(y_phys_test - y_test))

print(f"MAE of data-driven model trained on biased data: {mae_biased:.4f} ms")
print(f"MAE of physics baseline on test set: {mae_phys:.4f} ms")

plt.figure(figsize=(8,5))
#plt.scatter(balanced_test['fiber_distance_km'], y_test, label='True Latency', alpha=0.4)
#plt.plot(balanced_test['fiber_distance_km'], y_pred_biased, label='Data-driven Biased Model', color='orange')
#plt.scatter(balanced_test['fiber_distance_km'], y_phys_test, label='Physics Baseline', color='gray')

plt.scatter(balanced_test['geo_distance_km'], y_test, label='True Latency', alpha=0.4)
plt.plot(balanced_test['geo_distance_km'], y_pred_biased, label='Data-driven Biased Model', color='orange')
plt.scatter(balanced_test['geo_distance_km'], y_phys_test, label='Physics Baseline', color='red')
plt.xlabel("Fiber Distance (km)")
plt.ylabel("Latency (ms)")
plt.title("Generalization Performance on Balanced Test Set")
plt.legend()
plt.grid()
plt.show()

print("""
Data-driven model trained on biased data generalizes poorly to longer distances.
Physics baseline remains consistent and generalizes better even without training.
""")
```

```python
# -- 4. Add Realistic Noise and Systematic Biases (Enhanced) --

print("\n4. Noise and Systematic Bias Injection (Enhanced)")

df_noisy = df.copy()

# Add a non-linear region-based bias: certain lat/lon regions get extra bias
def complex_bias(row):
    lat_bias = 5.0 if (row['lat1'] > 20 and row['lon1'] < -100) else 0
    #lon_bias = 4.0 * np.sin(row['lon2'] / 50)  # non-linear by lon2
    lon_bias = 2.0 * np.sin(row['lon2'] / 50)
    return lat_bias + lon_bias

df_noisy['structured_bias'] = df_noisy.apply(complex_bias, axis=1)

# Add random unstructured congestion to ~50% of points
congestion_noise = np.zeros(len(df_noisy))
indices = np.random.choice(len(df_noisy), size=int(0.5 * len(df_noisy)), replace=False)
congestion_noise[indices] = np.random.normal(0, 10.0, size=len(indices))

# Add measurement noise
measurement_noise = np.random.normal(0, 3.5, size=len(df_noisy))

# Final noisy latency
df_noisy['measured_latency_ms_noisy'] = (
    df_noisy['measured_latency_ms'] +
    df_noisy['structured_bias'] +
    congestion_noise +
    measurement_noise
)

# Inputs
X_noisy = df_noisy[['geo_distance_km']].values

# Train data-driven model on clean data
model_dd_noisy = LinearRegression()
model_dd_noisy.fit(X_noisy, df_noisy['measured_latency_ms'])  # trained on clean labels

# Predict on noisy data
dd_pred = model_dd_noisy.predict(X_noisy)
physics_pred = df_noisy['ideal_latency_ms'].values 
y_noisy = df_noisy['measured_latency_ms_noisy'].values

# Compute residuals
residuals_dd_noisy = y_noisy - dd_pred
residuals_phys_noisy = y_noisy - physics_pred

# Variance comparison
var_dd_noisy = np.var(residuals_dd_noisy)
var_phys_noisy = np.var(residuals_phys_noisy)

print(f"Residual variance data-driven noisy: {var_dd_noisy:.3f}")
print(f"Residual variance physics baseline noisy: {var_phys_noisy:.3f}")

# Plot residual distributions
plt.figure(figsize=(8,4))
plt.hist(residuals_dd_noisy, bins=50, alpha=0.7, label='Data-Driven Residuals (Noisy)', color='orange')
plt.hist(residuals_phys_noisy, bins=50, alpha=0.7, label='Physics Residuals (Noisy)', color='green')
plt.legend()
plt.title("Residuals Distribution with Structured + Random Noise")
plt.xlabel("Residual (ms)")
plt.ylabel("Count")
plt.grid()
plt.show()

print("""
This enhanced noise simulation introduces non-linear, region-specific biases and strong congestion effects.
The physics-informed model — grounded in first principles — remains robust, while the data-driven model is misled by patterns it never saw during training.
Residual variance confirms the superior stability of physics-informed predictions under realistic, imperfect conditions.
""")
```
